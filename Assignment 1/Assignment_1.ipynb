```python
def max_count(arr):
    dict = {}
    for i in arr:
        if i in dict:
            dict[i] += 1
        else:
            dict[i] = 1
        
    max_element = max(dict, key=dict.get)
    return max_element 

x= np.array([1,2,3,4,5,1,2,1,1,1]) 
print(max_count(x))  

y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3])
print(max_count(y))
```

```python

def img_to_array(path):
    img = cv2.imread(path)
    
    if img is None:
        print("Error: Image not found at the given path.")
        return
    
    # Check image type
    if len(img.shape) == 2:  # Grayscale image
        print("Image is Grayscale.")
        np.savetxt("image_data.txt", img, fmt="%d")
    elif len(img.shape) == 3:  # RGB (actually BGR in OpenCV)
        print("Image is RGB.")
        # Save each channel separately
        for i, color in enumerate(['B', 'G', 'R']):
            np.savetxt(f"image_data_{color}.txt", img[:, :, i], fmt="%d")
    else:
        print("Unsupported image format.")

img_to_array("background.jpg")
```

```python


B = np.loadtxt("image_data_B.txt", dtype=np.uint8)
G = np.loadtxt("image_data_G.txt", dtype=np.uint8)
R = np.loadtxt("image_data_R.txt", dtype=np.uint8)
img_bgr = cv2.merge((B, G, R))

img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis('off')

```
